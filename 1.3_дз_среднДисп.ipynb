```python
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
a = []
for i in range(3):
    a.append(list(np.random.randint(1, 21, 10)) + list(np.random.randint(40, 101, 2)))

fig = make_subplots(rows=2, cols=3, subplot_titles=["Histogram 1", "Histogram 2", "Histogram 3"], vertical_spacing=0.05)

for i in range(3):
    fig.add_trace(go.Histogram(x=a[i], nbinsx=22), row=1, col=i+1)
    fig.add_trace(go.Scatter(x=list(range(len(a[i]))), y=a[i]), row=2, col=i+1)
    fig.update_layout(height=600, width=800, margin=dict(l=0, r=0, t=0, b=0), showlegend=False)
fig.show()
```

```python
import numpy as np
import plotly.graph_objects as go
a = []  # значение 1 - считаем за значение орла
figs = []  # Список для хранения отдельных графиков

for i in range(1):
    a.append(list(np.random.randint(1, 21, 10)) + list(np.random.randint(40, 101, 2)))
    figs.append(go.Figure(go.Histogram(x=a[i], nbinsx=22)))
    figs.append(go.Figure(go.Scatter(x=list(range(len(a[i]))), y=a[i])))

for fig in figs:
    fig.update_layout(height=200, width=400, margin=dict(l=0, r=0, t=0, b=0))
    fig.show()
```

```python
Столбцы в наборе данных px.data.gapminder() представляют различные показатели для каждой страны и года. Вот их описание:
country:  Название страны.
continent: Континент, к которому принадлежит страна.
year: Год, к которому относятся данные.
lifeExp: Ожидаемая продолжительность жизни в годах.
pop: Население страны.
gdpPercap: ВВП на душу населения (в долларах США).
iso_alpha: Код страны в формате ISO alpha-2 (двухбуквенный код страны).
iso_num: Числовой код страны в формате ISO (трехбуквенный код страны).
```
